```python
column = 'cast'
df[df[column].apply(lambda x: ',' in str(x))]
```

```python
# unnesting of the data
cast_df = df[['title', 'cast']]
cast_df["unnested_cast"] = cast_df['cast'].apply(lambda x: str(x).split(","))
cast_df= cast_df.explode("unnested_cast")
cast_df.head(10)
```

```python
column = "country"
df[df[column].apply(lambda x: ", " in str(x))]
```

```python
country_df = df[['title', 'country']]
country_df["unnested_country"] = country_df['country'].apply(lambda x: str(x).split(","))
country_df= country_df.explode("unnested_country")
country_df.head(10)
```

```python
#handling NAN values
df['director'] = df['director'].fillna(df['director'].mode()[0])
df['director']
```

```python
df['cast'] = df['cast'].fillna(df['cast'].mode()[0])
df['cast']
```

```python
df['country'] = df['country'].fillna(df['country'].mode()[0])
df['country']
```

```python
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
df['rating']
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['date_added'] = df['date_added'].fillna(df['date_added'].mode()[0])
df['date_added']
```

```python
# creating season and minutes column separately
df['seasons'] = df.loc[df['duration'].str.contains('Season', na=False), 'duration']
df['seasons'] = df['seasons'].fillna('unknown season')
df['seasons']
```

```python
df['minutes'] = df.loc[df['duration'].str.contains('min', na=False), 'duration']
df['minutes'] = df['minutes'].fillna(0)
df['minutes']
```

```python
#1. What type of content is available in top 10 different countries?
top_countries = df['country'].value_counts().head(10).index
df_filtered = df[df['country'].isin(top_countries)]
df_stacked_plot = pd.crosstab(index=df_filtered['country'], columns=df_filtered['type'])
df_stacked_plot
```

```python
df_stacked_plot.plot(kind='bar', stacked=True)
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Content Type by Country')
plt.show()
```

```python
#2. How has the number of movies released per year changed over the last 20-30 years?
yearly_counts=df['release_year'].value_counts().sort_values()
yearly_counts
```

```python
yearly_counts.plot(kind='line')
plt.xlabel('Release_Year')
plt.ylabel('Count')
plt.title('Number of Movies Released Per Year')
plt.show()
```

```python
#3.Comparison of tv shows vs. movies
comparison_count = df['type'].value_counts()
comparison_count.plot(kind='pie', autopct= '%.2f%%')
plt.title('Comparison of TV Shows vs. Movies')
plt.show()
```

```python
#4. Analysis of actors/directors of different types of shows/movies.
top_directors = df['director'].value_counts().head(10)
top_directors.plot(kind='bar')
plt.xlabel('Director')
plt.ylabel('Count')
plt.title('Top 10 Directors')
plt.show()
```

```python
#5. Does Netflix has more focus on TV Shows than Movies in recent years
yearly_counts = df['release_year'].value_counts()
yearly_counts
```
